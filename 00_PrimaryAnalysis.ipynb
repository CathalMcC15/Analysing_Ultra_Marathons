### Converting Times to Minutes

The first requirement for pre-prcessing the split times was to convert all times in the data frame from hour time to decimal minutes. I use the formula as shown in the 'minutes' function below.

```python
def minutes(time_str):
    return pd.to_timedelta(time_str).total_seconds()/60
```

I only want to apply the function to the actual split times, so I filter the irrelevant columns...

```python
df_non_stage = df[[ 'Unnamed: 0'  ,  'Year', 
'bib'              ,
'name_mask'           ,
'team'           ,
'gender',
'category'        , 
'rank'             ,
'nationality'      
 ]]
```

```python
stages = df.filter(like='to').applymap(minutes)
times = df.filter(like='time').applymap(minutes)
```

However, it is still hard to determine how fast individual stages were ran. I want to find the time it actually took a runner to complete a stage. The fact that there were many columns with a lot of nulls, as they were stages not present in certain years, presented some problems. I couldnt use a simple diff function to subtract one column from the previous column. Instead, I wrote a function for the cell of each stage column to subtract from the preceding non-null cell of the same row, which was present in a stage column. This gave the difference between the stages, ie. How long it took a runner to complete that stage in decimal minutes. 

```python
df2 = stages.where(stages.isnull(), 
                stages.fillna(method='ffill', axis=1).diff(axis=1))
df2.update(stages.where(stages.notnull().cumsum(1).cumsum(1) == 1))
```

## Gender Pacing

Having analysed the participation rates, the next step was to find out how this reflected their performance. To begin, I examined the average pace for males and females, and found that males were marginally faster on average. 

```python
df2.groupby(['gender'])['AVG_Pace'].mean()
```

```python
gen = df2.groupby(['gender'])['AVG_Pace'].mean()
ax = gen.plot.bar(figsize=(16, 8), title='Average 1km Pace per Gender', fontsize=18, rot=0)
ax.set_ylabel('Pace', fontsize = 14)
ax.set_xlabel('Gender', fontsize = 14)
```

I examined this further, year by year. 

```python
df2.groupby(['Year', 'gender'])['AVG_Pace'].mean().unstack()
```

```python
gen = df2.groupby(['Year', 'gender'])['AVG_Pace'].mean().unstack()
ax = gen.plot.bar(figsize=(16, 8), title='Average 1km Pace per Year and Gender', fontsize=18)
ax.set_ylabel('Pace', fontsize = 14)
ax.set_xlabel('Gender', fontsize = 14)
```

```python
gen = df2.groupby(['Year', 'gender'])['AVG_Pace'].mean().unstack()
ax = gen.plot(figsize=(16, 8), fontsize=24, marker='o', ms=15)
ax.set_ylabel('Pace', fontsize = 28)
ax.set_xlabel('Year', fontsize = 28)
ax.legend(['Female', 'Male'], frameon=True, ncol=2, loc='upper right', fontsize = 26)
```

The pace of both genders has followed a nearly identical path. The average pace dropped from 2008 to 2011, by nearly 2 minutes per kilometre for women, and just over 1 minute for men. The minutes per kilometre then rises until 2016, before it falls in 2017. Several factors could attribute to these changes. Route is one of them. 2016 differs from 2017 in three stages; Courmayeur2, La Catogne and Tete aux Vents. 2015 follows the exact same route as 2016, so it is quite probable these stages are causing the slower pace. However, 2014 only differs from these two years in one stage, and that is Courmayeur to Courmayeur2. One stage is unlikely to cause such a drastic difference in pace. 
Another factor that could contribute to this is participation. If a greater number are people are running, there could be more non-elite participants who cause an increase in average pace. 2015 and 2016 are the two highest years in terms of participation, as we seen in Figure 5 above. Then as the participation decreases so does the average pace. This pattern is not reflected in every year however, as 2008 and 2011 have very similar participation rates, but differ greatly in terms of average pace. The routes of these stages are different in two places however, so route cannot be ruled out as a factor.

Despite many of the studies I researched stating that the optimum age to particpate in an ultramarathon is between 40 and 49, this graph does not seem to reflect that. On the majority of years, Youth Men (20-22) and Senior Men (23-39) have the lowest pacing averages. However, the particpation races need to be revised before any conclusions are made. Only 34 particpants are between the ages of 20 and 22. This would give the indication that there is not enough data to get a fair and balanced view in relation to the other categories. There were even less particpants between the ages of 70 and 79, which is probably the reason they are on average faster than Master Men 3 (60-69) and Youth Men (20-22) categories in 2014. Therefore, the Senior Men category, participants aged between 23 and 39, are the best performers in terms of average pace. There is enough data to compare them to the highest participant, Master Men 1, and they are on average faster than them every year. It is the optimum age category for men in the UTMB. 

```python
cate = df2.groupby(['Year', 'category'])['AVG_Pace'].mean().unstack()
ax = Females.plot(figsize=(16, 8), fontsize=24, marker='o', ms=15)
ax.set_ylabel('Pace', fontsize = 30)
ax.set_xlabel('Year', fontsize = 30)
ax.legend( frameon=True, ncol=2, loc='upper center', fontsize=20)
```

The fastest pace in the graph is by Youth Women (20-22). Again, the participation levels need to be revised. Only 2 women of this age bracket competed, and they were obviously very elite athletes, so it would be unfair to compare this against the rest of the categories. Masters Women 3 (60-69) is another category with low participation, which is probably the reason they were the fastest category in 2013. Overall, Senior Women (23-39) are the best performers, akin to the age of the men. Despite Masters Women 1 (40-49) being the highest participant, the race suits the younger group better. 

Therefore, the optimum age for both Males and Females for the UTMB is between the ages of 23 and 39.
